```python
import pandas as pd
import pandas as pd
```

```python
data=pd.read_csv("/content/student_marks.csv")
print(data.head())
```

```python
data.rename(columns={'Unnamed: 0': 'Name'}, inplace=True)
marks_cols = ['Maths', 'Physics', 'Chemistry', 'English', 'Biology', 'Economics', 'History', 'Civics']
data['Average'] = data[marks_cols].mean(axis=1)
# Display result
print(data[['Name', 'Average']])
```

```python
data[['Name', 'Average']].to_csv('/content/student_avg.csv', index=False)
print("Average marks saved to student_avg.csv")
```

```python
import requests
from bs4 import BeautifulSoup

base_url = "http://books.toscrape.com/catalogue/page-{}.html"

book1 = []   # ✅ define list

for page in range(1, 20):
    page_url = base_url.format(page)
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, "lxml")

    books = soup.find_all("h3")

    for book in books:
        title = book.a["title"]   # ✅ extract title
        book1.append(title)

# Print all collected titles
for t in book1:
    print(t)

```

```python
books = soup.find_all("article", class_="product_pod")
book1=[]
# Mapping string ratings to numbers
rating_dict = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    rating_text = book.p["class"][1]  # e.g., "Three"
    rating_num = rating_dict[rating_text]  # convert to number

    # Only print books with rating > 2
    if rating_num > 2:
      book1.append(title)


```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
df.info()
```

```python
# Replace empty strings or special markers with actual NaN
df.replace(['', ' ', 'NA', 'NaN', 'missing'], np.nan, inplace=True)
# Now check nulls
print(df.isna().sum())
```

I AM REMOVING CABIN BECAUSE IT CONTAIN MANY NULL VALUE


```python
df.drop('Cabin', axis=1, inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)


```

```python

df.head(10)
```

```python
cat_cols = ['Sex', 'Embarked']
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['Age'])
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("age_histogram.png")
plt.show()

```

```python
x=df["Age"]
y=df["Fare"]
plt.scatter(x,y)
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.savefig("fare_vs_age.png")    # <- saved in current folder
plt.show()
```

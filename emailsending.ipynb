```python
import smtplib
```

```python
smtp_object=smtplib.SMTP("smtp.gmail.com",587)
```

```python
smtp_object.ehlo()
```

```python
smtp_object.starttls()
```

```python
import getpass
email =getpass.getpass(" email : ")
password=getpass.getpass("enter the password ")
```

```python
smtp_object.login(email,password)
```

```python
from_add=email
to="ua8141@gmail.com"
subject="the test email from python "
body="hy today i am testing gmail using python "
msg = "Subject: " + subject + "\n\n" + body
```

```python
smtp_object.sendmail(from_add,to,msg)
```
